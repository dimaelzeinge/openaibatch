```python
from peft import PeftModel
from transformers import AutoModelForCausalLM

base_model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen2-Audio-7B-Instruct")
model = PeftModel.from_pretrained(base_model, "cenkerenozbek/EnglishPronunciationEvaluation")
```

```python
from transformers import AutoProcessor
import torch
import soundfile as sf

processor = AutoProcessor.from_pretrained("Qwen/Qwen2-Audio-7B-Instruct")

# Load an audio sample (mono, 16kHz recommended)
audio_path = "learner.wav"
speech, sr = sf.read(audio_path)
if sr != 16000:
    import librosa
    speech = librosa.resample(speech, orig_sr=sr, target_sr=16000)
```

```python
instruction = "Evaluate the English pronunciation of the following recording and give a score between 0 and 100, plus short feedback."
```

```python
inputs = processor(
    text=instruction,
    audios=speech,
    sampling_rate=16000,
    return_tensors="pt"
).to(model.device)
```

```python
with torch.no_grad():
    output_ids = model.generate(**inputs, max_new_tokens=200)
```
